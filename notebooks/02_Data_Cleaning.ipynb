# Data Cleaning

## Objective

The objective of this notebook is to clean the raw Spotify dataset by handling missing values, removing duplicate records, standardizing text fields, validating data types, and preparing the dataset for exploratory data analysis and recommendation model development.

```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv("../data/raw/spotify_tracks.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
#handling missing values
df.isnull().sum()
```

```python
df["artists"] = df["artists"].fillna("Unknown")
df["album_name"] = df["album_name"].fillna("Unknown")
df["track_name"] = df["track_name"].fillna("Unknown")
```

```python
#removing duplicates
df.duplicated().sum()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
#removing duplicate songs
df[df.duplicated(subset=["track_name", "artists"])]
```

```python
df.duplicated(subset=["track_name", "artists"]).sum()
```

```python
duplicates = df[df.duplicated(subset=["track_name", "artists"], keep=False)]

duplicates.sort_values(by=["track_name", "artists"]).head(20)
```

Songs with the same name and artist were inspected. Many belong to different albums, compilations, or versions. Therefore, only exact duplicate rows were removed, while these records were retained.

```python
#removing extra spaces
text_columns = [
    "artists",
    "album_name",
    "track_name",
    "track_genre"
]

for col in text_columns:
    df[col] = df[col].str.strip()
```

```python
#standardize the genres
df["track_genre"] = df["track_genre"].str.lower()
```
