# Feature Engineering

## Objective

The objective of this notebook is to prepare the cleaned Spotify dataset for the recommendation system by selecting useful features, preprocessing them, and creating a feature matrix that represents each song based on its audio characteristics.

**Why is Feature Engineering Needed?**

Recommendation system cannot compare songs using text like:

Song A
Song B
Song C

**Instead, it compares numerical features such as:**

* Danceability
* Energy
* Loudness
* Acousticness
* Tempo
* Valence

Each song becomes a vector like:
The recommendation algorithm compares these vectors.

```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import StandardScaler
```

```python
df = pd.read_csv("../data/processed/clean_music.csv")

df.head()
```

**Select Important Features**

Choose only the audio features useful for recommendations

```python
features = [
    "danceability",
    "energy",
    "loudness",
    "speechiness",
    "acousticness",
    "instrumentalness",
    "liveness",
    "valence",
    "tempo"
]

music_features = df[features]

music_features.head()
```

```python
music_features.isnull().sum()
```

**Feature Scaling**

Since tempo has values around 120–180 while danceability is between 0 and 1, we need to scale them.

```python
scaler = StandardScaler()

scaled_features = scaler.fit_transform(music_features)
```

```python
scaled_df = pd.DataFrame(
    scaled_features,
    columns=features
)

scaled_df.head()
```

```python
scaled_df.describe()
```

**Combine with Song Information**

We'll need song names and artists later.

```python
final_df = pd.concat(
    [
        df[
            [
                "track_id",
                "track_name",
                "artists",
                "album_name",
                "track_genre",
                "popularity"
            ]
        ],
        scaled_df
    ],
    axis=1
)

final_df.head()
```

```python
final_df.to_csv(
    "../data/processed/music_features.csv",
    index=False
)
```

```python
feature_df = pd.read_csv("../data/processed/music_features.csv")

feature_df.head()
```

## Why Feature Scaling?

The numerical features in the dataset have different ranges.

For example:

- Danceability ranges from 0 to 1.
- Tempo ranges roughly from 50 to 200.

If left unscaled, features with larger values (like tempo) would dominate similarity calculations.

StandardScaler standardizes each feature to have a mean of 0 and a standard deviation of 1, ensuring that all features contribute fairly to the recommendation system.

# Conclusion

The important audio features were selected and standardized using StandardScaler.

The processed feature matrix was saved as `music_features.csv`.

This dataset is now ready for building a content-based music recommendation system using cosine similarity.
