# Exploratory Data Analysis (EDA)

## Objective

The objective of this notebook is to explore the cleaned Spotify dataset using statistical summaries and visualizations. The analysis aims to identify trends, distributions, relationships between features, and insights that will support the development of a music recommendation system.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.style.use("ggplot")
import seaborn as sns
```

```python
df = pd.read_csv("../data/processed/clean_music.csv")
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

The cleaned dataset contains 114000 songs and 20 features. Most numerical audio features are already normalized between 0 and 1.

```python
plt.figure(figsize=(8,5))

plt.hist(df["popularity"], bins=20)

plt.title("Distribution of Song Popularity")
plt.xlabel("Popularity")
plt.ylabel("Number of Songs")

plt.show()
```

Most songs have medium popularity scores, while only a small percentage of songs are extremely popular.

```python
top_genres = df["track_genre"].value_counts().head(10)

plt.figure(figsize=(10,5))

top_genres.plot(kind="bar")
plt.title("Top 10 Music Genres")
plt.xlabel("Genre")
plt.ylabel("Number of Songs")
plt.xticks(rotation=45)
plt.show()
```

The above seen genres in the graph dominante the dataset

```python
top_artists = df["artists"].value_counts().head(10)

plt.figure(figsize=(10,5))

top_artists.plot(kind="bar")

plt.title("Top 10 Artists")

plt.xticks(rotation=45)

plt.show()
```

The above graph shows us the top 10 artists in the given dataset.

```python
explicit_counts = df["explicit"].value_counts()

plt.figure(figsize=(5,5))

plt.pie(
    explicit_counts,
    labels=explicit_counts.index,
    autopct="%1.1f%%"
)

plt.title("Explicit vs Non-Explicit Songs")

plt.show()
```

Most songs in the dataset are non-explicit.

```python
plt.figure(figsize=(8,5))

plt.hist(df["danceability"], bins=20)

plt.title("Danceability Distribution")

plt.xlabel("Danceability")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["energy"], bins=20)

plt.title("Energy Distribution")

plt.xlabel("Energy")

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["tempo"], bins=20)

plt.title("Tempo Distribution")

plt.xlabel("Tempo")

plt.show()
```

```python
#average popularity by genre

genre_popularity = (
    df.groupby("track_genre")["popularity"]
      .mean()
      .sort_values(ascending=False)
      .head(10)
)

plt.figure(figsize=(10,5))

genre_popularity.plot(kind="bar")

plt.title("Average Popularity by Genre")

plt.xticks(rotation=45)

plt.show()
```

From the above graph we can observe that "pop-film" genre has the highest popularity.

**Average Duration**

```python
df["duration_minutes"] = df["duration_ms"] / 60000
```

```python
plt.figure(figsize=(8,5))

plt.hist(df["duration_minutes"], bins=20)

plt.title("Song Duration")

plt.xlabel("Minutes")

plt.show()
```

**Correaltion Analysis**

```python
numeric_df = df.select_dtypes(include=["int64", "float64"])

corr = numeric_df.corr()

plt.figure(figsize=(12, 8))

sns.heatmap(
    corr,
    annot=True,
    cmap="coolwarm",
    fmt=".2f",
    linewidths=0.5
)

plt.title("Correlation Heatmap of Numerical Features")

plt.show()
```

### Observation

- Features with correlation values close to **1** have a strong positive relationship.
- Features with correlation values close to **-1** have a strong negative relationship.
- Features with values near **0** have little or no linear relationship.
- Audio features such as **energy**, **loudness**, **danceability**, and **valence** show relationships that can be useful for recommending songs with similar characteristics.

**Top 10 Most Popular Songs**

```python
top_songs = df.sort_values("popularity", ascending=False).head(10)

top_songs[["track_name","artists","popularity"]]
```

**Top Albumns**

```python
top_albums = df["album_name"].value_counts().head(10)

plt.figure(figsize=(10,5))

top_albums.plot(kind="bar")

plt.title("Top Albums")

plt.xticks(rotation=45)

plt.show()
```

# Key Insights

- The dataset contains songs from a wide variety of artists and genres.
- Most songs have medium popularity scores.
- Danceability and energy are concentrated around moderate to high values.
- Certain genres dominate the dataset.
- Audio features such as danceability, energy, and tempo can be useful for building a recommendation system.
